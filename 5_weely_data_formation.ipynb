```python
import pandas as pd
# Assuming 'hukumu_score' is your DataFrame
# Group by 'User - JobSeekerId  Code' and calculate the average of 'Meeting  Score'
average_scores = hukumu_score.groupby('User - JobSeekerId__code')['Meeting__score'].mean()
# Reset the index to make it a DataFrame
average_scores = average_scores.reset_index()
# Rename the columns for clarity
average_scores.columns = ['user_code', 'hukumu_score']
# Display the result
print(average_scores)
```
